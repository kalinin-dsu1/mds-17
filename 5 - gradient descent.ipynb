```python
import numpy as np
from sympy import *
from scipy.optimize import approx_fprime
from scipy.optimize import minimize
```

### Уровень 0:


Посчитайте частные производные функций:

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$

```python
def func(x):
    "Coordinate vector `x` should be an array of size two."
    y = x[1]
    x = x[0]
    
    return 2*x**2*y**3 + 1/x + y**2*x + 7

x = np.ones(2)
eps = np.sqrt(np.finfo(float).eps)

approx_fprime(x, func, [eps, eps])
```

А для этой функции также градиент в точке $(1,2)$

```python
x = np.array([1,2])
approx_fprime(x, func, [eps, eps])
```

2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

```python
def func(x):
    "Coordinate vector `x` should be an array of size two."
    y = x[1]
    x = x[0]
    
    return x**2*y - sin(x*y) + cos(x**2) + 6*y

x = np.ones(2)
eps = np.sqrt(np.finfo(float).eps)

approx_fprime(x, func, [eps, eps])
```

### Уровень 1:
Градиентный спуск своими руками:

```python
def f(x):
    return (x[0]**2) + (x[1]**2)

x = np.array([100, 200])
learning_rate = 0.1
eps = np.sqrt(np.finfo(float).eps)


for i in range(100):
    # Calculate gradient
    grad = approx_fprime(x, f, [eps, eps])   
    
    # Update x with gradient
    x = x - grad * learning_rate

print("Минимум достигается в: ", x)
print("Значение функции в минимуме: ", f(x))
```

Проверка с помощью встроенной функции numpy:

```python
x = np.array([100, 200])
res = minimize(f, x, method='Nelder-Mead', options={'xtol':1e-8, 'disp':True})

print("Минимум достигается в: ", res.x)
print("Значение функции в минимуме: ", f(res.x))
```
