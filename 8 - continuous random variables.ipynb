```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
legend = [];
x = np.linspace(stats.norm.ppf(0.01), stats.norm.ppf(0.99), 100)

for scale in [0.2, 0.4, 0.6, 0.8, 1]:
    plt.plot(x, stats.norm.pdf(x, scale=scale))
    legend.append('scale = ' + str(scale))

plt.legend(legend)
plt.show()
```

```python
legend = [];
x = np.linspace(stats.expon.ppf(0.01, scale=scale), stats.expon.ppf(0.99, scale=scale), 100)

for scale in [0.2, 0.4, 0.6, 0.8, 1]:
    plt.plot(x, stats.expon.pdf(x, scale=scale))
    legend.append('scale = ' + str(scale))

plt.legend(legend)
plt.show()
```

```python
legend = [];


for df, scale in [(2,1),(5,1),(100,1),(2,0.8),(2,1.2)]:
        x = np.linspace(stats.t.ppf(0.01, df), stats.t.ppf(0.99, df), 100)
        plt.plot(x, stats.t.pdf(x, df, scale=scale))
        legend.append('df = ' + str(df) + ' scale = ' + str(scale))

plt.legend(legend)
plt.show()
```

Непрерывная случайная величина задана плотностью распределения:

$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$

Найдите C, математическое ожидание $\xi$ а также вероятность попадания  в отрезок $[3.5;7]$, т.е. P(3.5 &lt; $\xi$ &lt; 7).

**Решение**

Плотность распределения на промежутке [0;5] почти всюду постоянна, значит мы имеем дело с непрерывным равномерным распределением на отрезке [a,b]

$f(\xi) = 1 / (b - a), \xi \in [0;5]$

$C = 1 / (b - a) = 0.2$

$X \in [0;5]$

$P(3.5 \le X \le 7) = P(3.5 \le X \le 5) = 3.5 / 5 = 0.3$

Плотность вероятности почти всюду определена, значит

$E(\xi) = 0$

- Задание 3

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент.

```python
import math

def shannon(p):
    i = 0
    for x in p:
        i += x*math.log2(x)
    return -i

sum = (54 + 4 + 28)
red = 54 / sum
yellow = 4 / sum
green =  28 / sum

print(shannon([red, yellow, green]))
```

 А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
any = 1/3

print(shannon([any, any, any]))
```
