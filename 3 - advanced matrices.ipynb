### Придумайте матрицы, которые обладают следующими свойствами:

#### Диагональная, ортогональная, но не единичная

```python
A = np.array([
    [1, 0],
    [0, -1]
])

A @ np.matrix.transpose(A)
```

#### Верхнетреугольная, ортогональная

```python
A @ np.matrix.transpose(A)
```

#### Нижнетреугольная, ортогональная

```python
A @ np.matrix.transpose(A)
```

#### Матрица, у которой определитель = 1, и все элементы не равны 0

```python
A = np.array([
    [2, 1],
    [1, 1]
])

np.linalg.det(A)
```

### Найдите собственные значения и собственные вектора матриц:

```python
m = np.array([[2, 2],
              [1, 3]])
```

**Характеристическое уравнение матрицы:**

(2 - a)*(3 - a) - 2 * 1 = 0

a^2 - 5a + 4 = 0

Корни 1, 4 (они же - собственные значения)

**Подставляем первое значение, находим коофициэнты преобразования и первый собственный вектор**

1x + 2y = 0
1x + 2y = 0

x = -2y

Первый собственный вектор (-2, 1)

**Подставляем второе значение, находим коофициэнты преобразования и второй собственный вектор**

-2x + 2y = 0
x - y = 0

x = y

Второй собственный вектор (1, 1)

Проверяем:

```python
np.linalg.eig(m)
```

### Найдите спектральное разложение матрицы

```python
m = np.array([[1,2],
              [2,3]])
```

```python
eig = np.linalg.eig(m)

s_inv = eig[1]
d = np.diag(eig[0])
s = np.matrix.transpose(s_inv)

print(s_inv @ d @ s)
```

### Постройте зависимость качества восстановления рукописных цифр (MNIST) в зависимости от различного количества компонент PCA.

```python
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

plot_x = []
plot_y = []
for n_components in range(1, 65):
    pca = PCA(n_components = n_components)
    transformed = pca.fit_transform(digits.data)
    inversed = pca.inverse_transform(transformed)
    plot_x.append(n_components)
    plot_y.append(mean_squared_error(digits.data, inversed))

plt.plot(plot_x, plot_y)
```
