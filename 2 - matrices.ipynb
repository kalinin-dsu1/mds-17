```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Создайте матрицу 3 на 3, заполненную числами от 1 до 9

```python
matrix = np.arange(1, 10).reshape(3, 3)
matrix
```

#### Прибавьте к ней матрицу с единицами на диагонале

```python
matrix = matrix + np.eye(3, 3)
matrix
```

#### Умножьте полученную матрицу на вектор (1, 2, 3)

```python
multiplied_matrix = matrix @ [1,2,3]
multiplied_matrix
```

#### Посчитайте обратную матрицу к полученной в результате прибавления матрице

```python
inv_matrix = np.linalg.inv(matrix)
inv_matrix
```

#### Дан объект в 2d пространстве:

```python
A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
]

A = np.array(A)

def visualise(A):
    x = A[:,0]
    y = A[:,1]
    plt.plot(x, y)
    plt.ylim([-200, 500])
    plt.xlim([-200, 500])

visualise(A)
```

#### Уменьшить объект в два раза

```python
B = A @ (np.eye(3, 3) * 0.5)

visualise(A)
```

```python
visualise(B)
```

#### Повернуть на 130 градусов

```python
theta = np.radians(130)
c, s = np.cos(theta), np.sin(theta)
R = np.array([[c, -s, 0], [s, c, 0], [0, 0, 1]])
C = A @ R

visualise(A)
```

```python
visualise(C)
```

#### Отразить объект относительно прямой y=x

```python
D = A @ np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])

visualise(A)
```

```python
visualise(D)
```

#### Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)

```python
shift_matrix = np.eye(3,3)
shift_matrix[2][0] = 200
shift_matrix[2][1] = 300

scale_matrix = np.eye(3,3)
scale_matrix[0][0] = 0.5
scale_matrix[1][1] = 1.2


E = A @ shift_matrix @ scale_matrix

visualise(A)
```

```python
visualise(E)
```
