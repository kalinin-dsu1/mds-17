```python
from sympy import *
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

# Уровень 0:

Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

#### Используя sympy:

```python
x = symbols('x')

gtg = x**5 + 4*sin(2*x) + cos(3*x + 3)

first = diff(gtg).evalf(subs={x:1})
second = diff(diff(gtg)).evalf(subs={x:1})

print("Первая производная "+str(first))
print("Вторая производная "+str(second))
```

#### Используя scipy:

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)

first = derivative(f, 1, dx=1e-6)
second = derivative(f, 1, dx=1e-6, n=2)

print("Первая производная "+str(first))
print("Вторая производная "+str(second))
```

# Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5]

```python
def f(x):
    return sin(sin(sin(sin(sin(2*x+1)))))

x = np.linspace(-5, 5)
fx = [f(xx) for xx in x]   
f1x= [derivative(f, xx, dx=1e-6) for xx in x]
f2x= [derivative(f, xx, dx=1e-6, n=2) for xx in x]


plt.plot(x, fx, label='Function')
plt.plot(x, f1x, label='Derivative 1')
plt.plot(x, f2x , label='Derivative 2')
plt.legend()
plt.show()
```
